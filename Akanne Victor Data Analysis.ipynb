## 1. What is the total revenue/sales?

```python
#Make a new column for revenue
final_data["revenue"] = final_data["quantity"] * final_data["price"]

#Adding up all the revenue
total_revenue = final_data["revenue"].sum()

print(f"Total Revenue: ${total_revenue:,.2f}")
#the comma in the print help in adding large numbers and the,
#2f is rounding to 2decimal place in a float.
```

## 2. Find the total quantity sold.

```python
#Sum of the total quantity sold
total_quantity = final_data["quantity"].sum()

print(f"Total Quantity Sold: {total_quantity}")
```

## 3. Find the total orders.

```python
#Count the number of unique order_ids
total_orders = final_data["order_id"].nunique()

print(f"Total Orders: {total_orders}")
#the nunique count the number of different order_id are in the final_data.
```

## 4. How many pizza types do they sell?

```python
#Count unique pizza types
total_pizza_types = final_data["pizza_type_id"].nunique()

print(f"Total Pizza Types: {total_pizza_types}")
#the nunique count how many different pizza are in the final_data.
```

## 5. Find the average price of the pizzas.

```python
#Calculate average pizza price
average_price = final_data["price"].mean()

print(f"Average Pizza Price: ${average_price:.2f}")
#2f is rounding to 2decimal place in a float.
```

## 6. What are the peak hours of sales?

```python
final_data.info()
```

```python
#Convert the order_details_id column to datetime
final_data["order_details_id"] = pd.to_datetime(final_data["order_details_id"])

#Extract the hour from the datetime
final_data["hour"] = final_data["order_details_id"].dt.hour

#Count the number of orders per hour
orders_per_hour = final_data.groupby("hour")["order_id"].nunique()

#Find the hour(s) with the highest order count
peak_hour = orders_per_hour.idxmax()
max_orders = orders_per_hour.max()

print(f"Peak Hour: {peak_hour}:00 with {max_orders} orders")
```

## 7. Find the total sales made on each day of the week. Which day of the week is when sales are made the most?

```python
#Add a new column for the day of the week
final_data["day_of_week"] = final_data["order_details_id"].dt.day_name()

#Group by day of the week and sum the revenue
sales_by_day = final_data.groupby("day_of_week")["revenue"].sum()

#Find the day with the highest sales8.Find the top 5 bestselling pizzas.
top_day = sales_by_day.idxmax()
top_sales = sales_by_day.max()

print("Total Sales by Day of the Week:")
print(sales_by_day)
print(f"\nTop Sales Day: {top_day} with ${top_sales:,.2f}")
```

## 8. Find the top 5 bestselling pizzas.

```python
#Group by pizza and sum quantity sold
bestsellers = final_data.groupby("pizza_id")["quantity"].sum()

#Sort from highest to lowest
bestsellers = bestsellers.sort_values(ascending=False)

#Get the top 5
top_5_pizzas = bestsellers.head(5)

print("Top 5 Bestselling Pizzas:")
print(top_5_pizzas)
```

## 9. Find the sales made in each month. Any trend noticeable?

```python
#Create a column for the month name
final_data["month"] = final_data["order_details_id"].dt.month_name()

#Sum revenue by month
monthly_sales = final_data.groupby("month")["revenue"].sum()

#To see months in calendar order, sort by month number
final_data["month_num"] = final_data["order_details_id"].dt.month
monthly_sales_sorted = final_data.groupby(["month_num", "month"])["revenue"].sum().sort_index()

print("Monthly Sales:")
print(monthly_sales_sorted)
```

## 10. Are there pizza types that are not doing well on the menu?

```python
#Group by pizza type and sum the quantity sold
pizza_sales = final_data.groupby("pizza_id")["quantity"].sum()

#Sort from lowest to highest
worst_selling = pizza_sales.sort_values(ascending=True)

print("Lowest Selling Pizzas:")
print(worst_selling.head(5))
```
